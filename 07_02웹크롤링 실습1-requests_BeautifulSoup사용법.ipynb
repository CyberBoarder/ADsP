```python
def text_clean(text):
    return text.strip().replace("\xa0", " ").replace("\n", " ").replace("\t", " ").replace("\r", " ").replace("|", " ")
    
```

for 문을 사용해서 dl 에서 자료를 한 개씩 꺼내서 book에 담고 그 안에 있는 dt, dd 태그에 있는 자료 추출

```python
books_info = []
for book in books.find_all("dl"):
    book_info = []
    book_info.append(text_clean(book.find("dt").get_text()))
    for contents in book.find_all("dd"):
        book_info.append(text_clean(contents.get_text()))
    books_info.append(book_info)
print(books_info)
```

### 제목과 책 소개는 1개의 리스트 데이터로, 저자, 출판사, 별점등은 각각의 리스트 요소로 만들고 특수문자 제거

```python
def clean(index, text):
    result = []
    if index != 2:
        result .append(text_clean(text).split())
    elif index == 2:
        result.append(text_clean(text))
    return result
```

```python
def clean(index, text):
    result = []
    if index != 2:
        result.append(text.strip().replace("\xa0", " ").replace("\n", " ").replace("\t", " ").replace("\r", " ").replace("|", " ").replace("  "," ").split())
    elif index == 2:
        result.append(text.strip().replace("\xa0", " ").replace("\n", " ").replace("\t", " ").replace("\r", " ").replace("|", " ").replace("  "," "))
    return result
```

```python
select_books = []
for items in book_list:
    book =[]
    title = clean(2, items.select_one("dt").get_text())
    book.append(title[0])
    for index, item in enumerate(items.select("dd")):
        item_contents = clean(index, item.get_text())
        book.append(item_contents[0])
    select_books.append(book)
print(select_books)
        
```
