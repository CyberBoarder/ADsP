```python
def age_adj(x):
    if x <10:
        return "Under 10s"
    elif 10 <= x < 20:
        return "10s"
    elif 20 <= x < 30:
        return "20s"
    elif 30 <= x < 40:
        return "30s"
    elif 40 <= x < 50:
        return "40s"
    elif 50 <= x < 60:
        return "50s"
    elif 60 <= x < 70:
        return "60s"
    elif 70 <= x < 80:
        return "70s"
    
    
```

```python
data['Age'] = data['Age'].apply(lambda x : age_adj(x))
```

```python
data
```

```python
data['Age'].value_counts()
```

# one-hot encoding 

```python
data = pd.get_dummies(data, columns=["Pclass", "Sex", "Age", "Embarked"], drop_first=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('Survived', axis = 1)
y = data['Survived']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=100)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=10, random_state= 100)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```

```python
for i in range(2,101):
    model = DecisionTreeClassifier(max_depth=i, random_state= 100)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print("max_depth= ", i, accuracy_score(y_test, pred))
#     print(confusion_matrix(y_test, pred))
#     print(classification_report(y_test, pred))
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state= 100)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```
